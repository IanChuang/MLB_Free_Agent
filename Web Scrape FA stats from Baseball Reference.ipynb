```python
# Check if all response are 200 
response_list = []
for i in range(len(FA_list)):
    file_FA = requests.get("https://www.baseball-reference.com/players/{}/{}.shtml".format(FA_list["bbrefID"][i][0], FA_list["bbrefID"][i]))
    response_list.append(file_FA)
response_num_list = []
for i in range(len(response_list)):
    response_num = response_list[i].status_code
    response_num_list.append(response_num)
np.unique(response_num_list)
```

```python
FA_stats = []
for i in range(len(FA_list)):
    if FA_list["Position"][i] in ["SP", "RP"]:
        file_FA = requests.get("https://www.baseball-reference.com/players/{}/{}.shtml".format(FA_list["bbrefID"][i][0], FA_list["bbrefID"][i]))
        html_FA = BeautifulSoup(file_FA.text, "html.parser")
        colname_table = html_FA.findAll("div", class_ = "overthrow table_container")
        stats_table = html_FA.findAll("tr", class_ = "full")
        colname = []
        for i in range(34):
            colname.append(colname_table[0].select("tr")[0].select("th")[i].get_text())
        first_row = [stats_table[0].select("th")[0].get_text()]
        for i in range(33):
            first_row.append(stats_table[0].select("td")[i].get_text())
        row_list = []
        for i in range(len(stats_table)):
            row = [stats_table[i].select("th")[0].get_text()]
            for j in range(33):
                row.append(stats_table[i].select("td")[j].get_text())
            row_list.append(row)
        df = pd.DataFrame(data = row_list, columns = colname)
        df[df.columns.drop(["Tm","Lg"])] = df[df.columns.drop(["Tm","Lg"])].apply(pd.to_numeric)
        # Make all columns except for Tm and Lg numeric 
    else:
        file_FA = requests.get("https://www.baseball-reference.com/players/{}/{}.shtml".format(FA_list["bbrefID"][i][0], FA_list["bbrefID"][i]))
        html_FA = BeautifulSoup(file_FA.text, "html.parser")
        colname_table = html_FA.findAll("div", class_ = "overthrow table_container")
        stats_table = html_FA.findAll("tr", class_ = "full")
        colname = []
        for i in range(28):
            colname.append(colname_table[0].select("tr")[0].select("th")[i].get_text())
        row_list = []
        for i in range(len(stats_table)):
            row = [stats_table[i].select("th")[0].get_text()]
            for j in range(27):
                row.append(stats_table[i].select("td")[j].get_text())
            row_list.append(row)
        df = pd.DataFrame(data = row_list, columns = colname)
        df[df.columns.drop(["Tm","Lg"])] = df[df.columns.drop(["Tm","Lg"])].apply(pd.to_numeric)
    FA_stats.append(df)
```

BA = H/AB    <br>
OBP = (Hits + Walks + Hit by Pitch) / (At Bats + Walks + Hit by Pitch + Sacrifice Flies) (baseball reference)   <br>
SLG = (Singles + Doubles x 2 + Triples x 3 + Home Runs x 4)/At Bats   <br>
OPS = OBP + SLG   <br>
Drop OPS+ because it's unattainable    <br>
