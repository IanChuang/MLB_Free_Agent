```python
def PitcherDF(url):
    file = requests.get(url + "statistics")
    html = BeautifulSoup(file.text, "html.parser")
    table = html.findAll(text = "Season")[0].parent.parent.parent.parent
    Season = list()
    Team = list()
    GP = list()
    GS = list()
    IP = list()
    H = list()
    R = list()
    ER = list()
    BB = list()
    K = list()
    HR = list()
    W = list()
    L = list()
    S = list()
    ERA = list()
    WHIP = list()
    WAR = list()
    Name = list()
    length = len(table.select("tr"))
    for i in range(1, length):
        Season.append(table.select("tr")[i].select("td")[0].get_text())
        Team.append(table.select("tr")[i].select("td")[1].get_text())
        GP.append(table.select("tr")[i].select("td")[2].get_text())
        GS.append(table.select("tr")[i].select("td")[3].get_text())
        IP.append(table.select("tr")[i].select("td")[4].get_text())
        H.append(table.select("tr")[i].select("td")[5].get_text())
        R.append(table.select("tr")[i].select("td")[6].get_text())
        ER.append(table.select("tr")[i].select("td")[7].get_text())
        BB.append(table.select("tr")[i].select("td")[8].get_text())
        K.append(table.select("tr")[i].select("td")[9].get_text())
        HR.append(table.select("tr")[i].select("td")[10].get_text())
        W.append(table.select("tr")[i].select("td")[11].get_text())
        L.append(table.select("tr")[i].select("td")[12].get_text())
        S.append(table.select("tr")[i].select("td")[13].get_text())
        ERA.append(table.select("tr")[i].select("td")[14].get_text())
        WHIP.append(table.select("tr")[i].select("td")[15].get_text())
        WAR.append(table.select("tr")[i].select("td")[16].get_text())
        Name.append(table.parent.parent.parent.parent.parent.select("h1")[0].get_text().strip())
    df_Dict = {"Season":Season, "Team":Team, "GP":GP, "GS":GS, "IP":IP, "H":H, "R":R, "ER":ER, "BB":BB, "K":K, "HR":HR, "W":W, "L":L, "S":S, "ERA":ERA, "WHIP":WHIP, "WAR":WAR, "Name":Name}
    df = pd.DataFrame(df_Dict)
    return df
```

```python
def RelieverDF(url):
    file = requests.get(url + "statistics")
    html = BeautifulSoup(file.text, "html.parser")
    table = html.findAll(text = "Season")[0].parent.parent.parent.parent
    Season = list()
    Team = list()
    Games = list()
    IP = list()
    H = list()
    R = list()
    ER = list()
    BB = list()
    K = list()
    HR = list()
    W = list()
    L = list()
    S = list()
    ERA = list()
    WHIP = list()
    WAR = list()
    Name = list()
    length = len(table.select("tr"))
    for i in range(1, length):
        Season.append(table.select("tr")[i].select("td")[0].get_text())
        Team.append(table.select("tr")[i].select("td")[1].get_text())
        Games.append(table.select("tr")[i].select("td")[2].get_text())
        IP.append(table.select("tr")[i].select("td")[3].get_text())
        H.append(table.select("tr")[i].select("td")[4].get_text())
        R.append(table.select("tr")[i].select("td")[5].get_text())
        ER.append(table.select("tr")[i].select("td")[6].get_text())
        BB.append(table.select("tr")[i].select("td")[7].get_text())
        K.append(table.select("tr")[i].select("td")[8].get_text())
        HR.append(table.select("tr")[i].select("td")[9].get_text())
        W.append(table.select("tr")[i].select("td")[10].get_text())
        L.append(table.select("tr")[i].select("td")[11].get_text())
        S.append(table.select("tr")[i].select("td")[12].get_text())
        ERA.append(table.select("tr")[i].select("td")[13].get_text())
        WHIP.append(table.select("tr")[i].select("td")[14].get_text())
        WAR.append(table.select("tr")[i].select("td")[15].get_text())
        Name.append(table.parent.parent.parent.parent.parent.select("h1")[0].get_text().strip())
    df_Dict = {"Season":Season, "Team":Team, "Games":Games, "IP":IP, "H":H, "R":R, "ER":ER, "BB":BB, "K":K, "HR":HR, "W":W, "L":L, "S":S, "ERA":ERA, "WHIP":WHIP, "WAR":WAR, "Name":Name}
    df = pd.DataFrame(df_Dict)
        # Because the format on the website is messed up. All entries should be moved one step left. Calculate cummulative WAR
    df.iloc[-1,3] = df.iloc[-1, 4]
    df.iloc[-1,4] = df.iloc[-1, 5]
    df.iloc[-1,5] = df.iloc[-1, 6]
    df.iloc[-1,6] = df.iloc[-1, 7]
    df.iloc[-1,7] = df.iloc[-1, 8]
    df.iloc[-1,8] = df.iloc[-1, 9]
    df.iloc[-1,9] = df.iloc[-1, 10]
    df.iloc[-1,10] = df.iloc[-1, 11]
    df.iloc[-1,11] = df.iloc[-1, 12]
    df.iloc[-1,12] = df.iloc[-1, 13]
    df.iloc[-1,13] = df.iloc[-1, 14]
    df.iloc[-1,14] = df.iloc[-1, 15]
    df.iloc[-1,15] = sum(pd.to_numeric(df.iloc[0:length-2,15]))
    return df
```

```python
def PlayerDF(input_url, input_position):
    length = len(input_url)
    df_output = list()
    for i in range(length):
        if input_position[i] == "SP":
            df = PitcherDF(input_url[i])
        elif input_position[i] == "RP":
            df = RelieverDF(input_url[i])
        else:
            df = BatterDF(input_url[i])
        df_output.append(df)
    return df_output
```
